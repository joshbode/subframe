```python
import subframe
subframe.plugins.enable()
```

```python
import pandas
import numpy
numpy.random.seed(1)

data = pandas.DataFrame({
    'a': numpy.random.randint(1, 5, 1000),
    'b': numpy.random.randint(1, 10, 1000),
    'x': numpy.random.randn(1000),
    'y': numpy.random.randn(1000)
})
```

```python
subframe.PivotTable(
    data, labels={'a': 'Foo', 'b': 'Bar'},
    options={'rows': ['Foo'], 'cols': ['Bar'], 'rendererName': 'Area Chart'}
)
```

```python
subframe.DataTable(data, labels={'index': 'Index', 'a': 'Foo', 'b': 'Bar'})
```
